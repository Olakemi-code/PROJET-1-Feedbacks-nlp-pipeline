## VISUALISATION DES SENTIMENTS

```python
st.subheader("Répartition des sentiments")
sentiment_counts = data['sentiment'].value_counts()
st.bar_chart(sentiment_counts)
```

## NUAGES DE MOTS PAR CLUSTER

```python
st.subheader("Nuage de mots par cluster")

clusters = sorted(data["cluster"].unique())
cluster_choisi = st.selectbox("Choisir un cluster", clusters)

text = " ".join(data[data["cluster"] == cluster_choisi]["cleaned_title"])
wordcloud = WordCloud(width=800, height = 400, background_color = "white").generate(text)

fig, ax = plt.subplots(figsize=(10,5))
ax.imshow(wordcloud, interpolation='bilinear')
ax.axis("off")
st.pyplot(fig)
```

## FILTRE PAR SENTIMENT 

```python
st.subheader("Filtrer les avis par sentiment")
selected_sentiment = st.radio("Sentiment", data["sentiment"].unique())

filtered_data = data[data["sentiment"] == selected_sentiment]
st.dataframe(filtered_data[["title", "numeric_rating", "cluster"]])
```

```python
st.download_button(
    label="Télécharger les avis filtrés",
    data=filtered_data.to_csv(index=False).encode('utf-8'),
    file_name=f'avis_{selected_sentiment}.csv',
    mime='text/csv'
)
```
